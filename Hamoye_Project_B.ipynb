```python
x1 = x [['Temperature in living room area', 'Temperature outside the building (north side)']].sample(30,random_state=1)
```

```python
sns.regplot(x='Temperature in living room area', y='Temperature outside the building (north side)', data=x1)
```

```python
#To normalize data to column scale usingthe min scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
nor_x1= pd.DataFrame(scaler.fit_transform(x1), 
columns=x1.columns)
fe_x = nor_x1.drop(columns=
['Temperature outside the building (north side)'])
temp_building = nor_x1['Temperature outside the building (north side)']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(fe_x, temp_building,
test_size=0.3, random_state=2)
```

```python
import sklearn
from sklearn import linear_model
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
#Fit the model to the training dataset
linear_model.fit(x_train,y_train)
#Obtain predictions
predicted_values=linear_model.predict(x_test)
predicted_values
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score,2)
```
